```python
print(data.info()) # To see the structure of the DataFrame
print(data.describe()) # Basic statistical details 
```

```python
# which country has the most number of players
top_countries = data['Nationality'].value_counts().head(1)
print(top_countries)
```

```python
#  bar chart of 5 top countries with the most number of players
import matplotlib.pyplot as plt
top_5_countries = data['Nationality'].value_counts().nlargest(5).index
plt.figure(figsize=(8,5))
plt.bar(top_5_countries, data['Nationality'].value_counts().nlargest(5).values)
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.title('top 5 Countries with Most Number of Players')
plt.xticks(rotation=45)
plt.show()
```

```python
# find player with highest salary
import pandas as pd
data = pd.read_csv(r"C:\Users\nafla\OneDrive\Desktop\data\fifa_data.csv")
df = pd.DataFrame(data)
df['right_extract'] = df['Wage'].str[-1:]
df['salary'] = df['right_extract'].str[1:]
highest_salary_player = df.loc[df['salary'].idxmax()]
print(f"the player with the highest salary is{highest_salary_player['Name']} ")
```

```python
#histogram to get the salary range of players
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plt.hist(data['Wage'], bins=20, color='blue', edgecolor='red')
plt.xlabel('Wage')
plt.ylabel('Frequency')
plt.title('Wage Range of Players')
plt.show()
```

```python
#find the tallest player
tallest_player = data.loc[data['Height'].idxmax()]
print(f"tallestplayer is{tallest_player['Name']}")
```

```python
# which club has the most number of players
top_club = data['Club'].value_counts().head(1)
print(top_club)
```

```python
# bar chart for preferred foot
preferred_foot = data['Preferred Foot'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(preferred_foot.index, preferred_foot.values)
plt.xlabel('Preferred Foot')
plt.ylabel('Number of Players')
plt.title('Most Preferred Foot Among Players')
plt.show()
           
```
